```python
dataset['education'] = dataset['education'].replace({'Graduate': 1, 'Not Graduate': 0})
dataset['loan_status'] = dataset['loan_status'].replace({'Approved': 1, 'Rejected': 0})
dataset['self_employed'] = dataset['self_employed'].replace({'Yes': 1, 'No': 0})
display(dataset.head())
```

```python
#drop whose column who has least impact
dataset = dataset.drop(['loan_id', 'self_employed', 'education', 'commercial_assets_value'], axis=1)
```

```python
print(dataset.columns)
```

```python
dataset['loan_status'] = dataset['loan_status'].replace({'Approved': 1, 'Rejected': 0})
display(dataset.head())
dataset.shape
```

```python
# Splitting dataset into independent and dependent features
X = dataset[['no_of_dependents', 'income_annum', 'loan_amount', 'loan_term','cibil_score', 'residential_assets_value', 'luxury_assets_value',
       'bank_asset_value']]
y = dataset['loan_status']
print(X.head())
```

```python
print(y.head())
```

```python
#Split Data into Training and Testing Sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
#Create and Train the Model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
#Predict on Training and Testing Data
from sklearn.metrics import accuracy_score
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
print("Training Accuracy:", train_accuracy)
print("Testing Accuracy:", test_accuracy)
```

```python
#report
from sklearn.metrics import classification_report, confusion_matrix
print("Classification Report:")
print(classification_report(y_test, y_test_pred))
conf_matrix = confusion_matrix(y_test, y_test_pred)
print("Confusion Matrix:\n", conf_matrix)
```

```python
sns.heatmap(conf_matrix, annot=True, cmap="Blues", fmt='g')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
plt.bar(['Training Accuracy', 'Testing Accuracy'], [train_accuracy, test_accuracy], color=['green', 'orange'])
plt.ylim(0, 1)
plt.title("Training vs Testing Accuracy")
plt.ylabel("Accuracy")
plt.show()
```
